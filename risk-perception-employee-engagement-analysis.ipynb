```python
import matplotlib.pyplot as plt
import pandas as pd

# Data for the questions
questions_data = {
    "Awareness of Organizational Risk Levels": {
        "labels": ["I don't know", "Very low extent", "Low extent", "Moderate", "High extent", "Very high extent"],
        "values": [0, 2, 3, 7, 21, 17],
    },
    "Perception of Organizational Risk Appetite": {
        "labels": ["I don't know", "Very low extent", "Low extent", "Moderate", "High extent", "Very high extent"],
        "values": [13, 0, 6, 19, 6, 6],
    },
    "Tolerance for Unethical Behavior": {
        "labels": ["I don't know", "Very low extent", "Low extent", "Moderate", "High extent", "Very high extent"],
        "values": [0, 2, 2, 7, 13, 26],
    },
    "Personal Engagement and Commitment": {
        "labels": ["I don't know", "Very low extent", "Low extent", "Moderate", "High extent", "Very high extent"],
        "values": [0, 0, 1, 7, 16, 26],
    },
    "Employment Agreement Type": {
        "labels": ["I am a contracted employee", "I am a permanent employee"],
        "values": [4, 46],
    },
}

def create_bar_chart(question, labels, values):
    plt.figure(figsize=(10, 6))
    bars = plt.bar(labels, values, color='skyblue')
    
    plt.xlabel('Responses', fontsize=12)
    plt.ylabel('Number of Respondents', fontsize=12)
    plt.title(f"{question} - Frequency of Responses", fontsize=14)
    
    # Rotate x-axis labels to prevent overlap
    plt.xticks(rotation=45, ha='right')
    
    # Add values above bars
    for i, bar in enumerate(bars):
        yval = bar.get_height()
        plt.text(i, yval + 0.5, str(yval), ha='center', fontsize=10)
    
    # Adjust y-axis limit to create space above bars
    plt.ylim(0, max(values) + 5)  # Add extra space above bars
    
    # Adjust layout to make sure everything fits well
    plt.tight_layout()

    # Save chart to PNG
    filename = f"{question.replace(' ', '_').replace('(', '').replace(')', '')}_BarChart.png"
    plt.savefig(filename, format='png')
    print(f"Bar chart saved as {filename}")
    
    plt.show()

def create_pie_chart(question, labels, values):
    plt.figure(figsize=(8, 8))
    plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
    plt.title(f"{question} - Percentage Distribution", fontsize=14)
    plt.tight_layout()

    # Save chart to PNG
    filename = f"{question.replace(' ', '_').replace('(', '').replace(')', '')}_PieChart.png"
    plt.savefig(filename, format='png')
    print(f"Pie chart saved as {filename}")
    plt.show()

# Generate the visualizations
for question, data in questions_data.items():
    print(f"Creating visualizations for {question}...")
    create_bar_chart(question, data['labels'], data['values'])
    create_pie_chart(question, data['labels'], data['values'])
```
